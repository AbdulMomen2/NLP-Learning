```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
import pandas as pd
nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_json("doj_press.json", lines=True)
```

```python
df.head()
```

##Preprosessing 

```python
df.shape
```

```python
df =df[df['topics'].str.len() != 0] #removing empty topics from column
df.head()
```

```python
df =df.head(1000)
df.shape
```

```python
def preprocess(text):
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)  
    return " ".join(filtered_tokens)
```

```python
df["contents_new"]= df.contents.apply(preprocess)
```
